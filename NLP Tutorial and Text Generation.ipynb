##  Word senses (for homonyms)

WordNet is a lexical database for the English language in the form of a semantic graph.

WordNet groups English words into sets of synonyms called synsets, provides short definitions and usage examples, and records a number of relations among these synonym sets or their members.

NLTK provides an interface to the WordNet API.

```python
from nltk.corpus import wordnet as wn
wn.synsets('human')
```

```python
human = wn.synsets('Human', pos=wn.NOUN)[0]
human.hyponyms()
```

```python
text = 'We have higher buildings, but shorter patience; we have wider motorways, but narrower perspectives. We spend more, but we have less; we buy more, but we are less satisfied. Our larger homes, but smaller families; more home appliances, but we have less time. More education, but less common sense; we have more knowledge, but less wisdom. More of our experts, but still more problems; we have more medicine, but less health. We consume a lot of alcohol and cigarettes, spend a lot of extravagant money, laugh very little, drive very fast, get angry very quickly, sit up very late, get up very tired, read very little, watch too much TV, and thank very rarely. We increased our assets, but decreased our values. We talk a lot, we love very little and we hate very often. We learned to make a living, but we could not learn to make a living. We added years to our lives, but we could not add years to life. We learned to go to the moon, but we have a problem crossing the street to meet our new neighbor. We conquered the Outer Space, but we could not conquer our inner world. We did bigger things, but we couldnt do better things. We cleaned the air, but we polluted our soul. We have ruled the atom, but we have not been able to prejudice. We write more, but we learn less. We plan more, we draw less. We learned to run, but we could not learn to wait.We build more computers to store more information, make more copies than ever, but we are less and less communicating. Time is now the fast food and slow digestion of food; big guys and small characters; its time for high profits and shallow relationships. Nowadays are days when two salaries enter, but more divorces, more fancy houses, but scattered homes. These days are fast travels, disposable diapers, moral values ​​destroyed, overnight relationships, obese bodies, and pills that can do anything from cheering to calming or even killing. We are at a time when everything is on display, but nothing is in the warehouses. At a time, technology can bring this letter to you, you can choose to either share this internality or press the delete button. Life is measured not by the number of breaths we take, but by the number of moments that take our breath away.'
#Respect George Carlin
from collections import defaultdict

def markov_chain(text):
    '''The input is a string of text and the output will be a dictionary with each word as
       a key and each value as the list of words that come after the key in the text.'''
    
    # Tokenize the text by word, though including punctuation
    words = text.split(' ')
    
    # Initialize a default dictionary to hold all of the words and next words
    m_dict = defaultdict(list)
    
    # Create a zipped list of all of the word pairs and put them in word: list of next words format
    for current_word, next_word in zip(words[0:-1], words[1:]):
        m_dict[current_word].append(next_word)

    # Convert the default dict back into a dictionary
    m_dict = dict(m_dict)
    return m_dict
```

```python
text = markov_chain(text)
```

```python
import random

def generate_sentence(chain, count=10):
    '''Input a dictionary in the format of key = current word, value = list of next words
       along with the number of words you would like to see in your generated sentence.'''

    # Capitalize the first word
    word1 = random.choice(list(chain.keys()))
    sentence = word1.capitalize()

    # Generate the second word from the value list. Set the new word as the first word. Repeat.
    for i in range(count-1):
        word2 = random.choice(chain[word1])
        word1 = word2
        sentence += ' ' + word2

    # End it with a period
    sentence += '.'
    return(sentence)
```
